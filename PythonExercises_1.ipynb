**Text Analysis Tool:** Create a Python program that reads a text file and performs various analyses on its content. The program should: Count the number of words, sentences, and paragraphs. Calculate the average word length. Identify the most common words and their frequencies.

```python
# The Natural Language Toolkit (NLTK) is a Phyton library that is primarily
# used for working with human language data, also known as natural language
# processing (NLP). NLTK provides a wide range of tools and resources for
# various tasks related to analyzing and processing text data.

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize # tokenization: process of splitting text into sentences and words.
# sent_tokenize: is used to tokenize text into sentences.
# word_tokenize: is used to tokenize text into words.
from nltk.probability import FreqDist
# FreqDist is used for frequency distribution analysis. Counting frequency of words in a text.

# Download NLTK data if not already downloaded
nltk.download('punkt') # Punkt tokenizer models, these models are used for tokenization tasks.

# Function to count words, sentences, and paragraphs
def count_stats(text):
    sentences = sent_tokenize(text)
    words = word_tokenize(text)
    paragraphs = text.split('\n\n')
    return len(words), len(sentences), len(paragraphs)

# Function to calculate the average word length
def average_word_length(text):
    words = word_tokenize(text)
    total_word_length = sum(len(word) for word in words)
    return total_word_length / len(words)

# Function to identify the most common words and their frequencies
def most_common_words(text, n=10):
    words = word_tokenize(text)
    fdist = FreqDist(words)
    return fdist.most_common(n)

# Main program
if __name__ == "__main__":
    filename = "Prueba1.txt"  # Replace with the path to your text file

    try:
        with open(filename, "r") as file:
            text = file.read()

        word_count, sentence_count, paragraph_count = count_stats(text)
        avg_word_length = average_word_length(text)
        common_words = most_common_words(text)

        print("Word count:", word_count)
        print("Sentence count:", sentence_count)
        print("Paragraph count:", paragraph_count)
        print("Average word length:", avg_word_length)
        print("\nMost common words:")
        for word, freq in common_words:
            print(f"{word}: {freq}")

    except FileNotFoundError:
        print(f"File '{filename}' not found.")
```

**Hangman Game:** Create a Python program that allows a user to play the game of Hangman. The program should randomly select a word from a predefined list, and the player must guess letters one at a time until they guess the word correctly or run out of attempts.

```python
import random

stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

# List of words for the game
word_list = ["abruptly", "buffalo", "cycle", "dizzying", "exodus", "flyby", "gossip", "injury"]

# Select a random word from the list
chosen_word = random.choice(word_list)
word_length = len(chosen_word)

end_of_game = False
lives = 6

#Create blanks
display = []
for _ in range(word_length):
  display += "_"

while not end_of_game:
  guess = input("Guess a letter: ").lower()

  if guess in display:
    print(f"You have alerady guessed {guess}")

  #Cheking guessed letter
  for position in range(word_length):
    letter = chosen_word[position]

    if letter == guess:
      display[position] = letter

  #Check if user is wrong
  if guess not in chosen_word:
    if guess not in display:
      print(f"{guess} is not in the word, you lose a life!")
    lives -= 1
    if lives == 0:
      end_of_game = True
      print("You lose!")

  #Joining all the elements in the list and turn it into a string.
  print(f"{' '.join(display)}")

  #Checking if user has got all letters.
  if "_" not in display:
    end_of_game = True
    print("You win!")


  #Stages
  print(stages[lives])
```

**Number Guessing Game:** Create a program that generates a random number between 1 and 100, and then asks the user to guess the number. The program should provide hints to the user if the guessed number is greater or less than the target number, and keep asking for a new guess until the user guesses correctly.

```python
import random

random_num = random.randint(1, 100) # Generating a random number between 1 and 100.

while True:
  guessed_num = int(input("Guess the number: ")) # The program asks the user to enter a number.

  if guessed_num == '0':           # This part of the code allows the user to quit the game
    print("You quit the game.")    # if she/he wants to.
    break

  guessed_num = int(guessed_num)

  # Checking if the user's guess is equal to the randomly generated target number.
  if guessed_num == random_num:
    print("You win!")
    break
  elif guessed_num > random_num:
    print("Ups, the number is smaller")
  else:
    print("Ups, the number is greater")
```
